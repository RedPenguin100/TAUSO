# ClinASO competitor feature

Populates the ClinASO RNase H1 PWM potency score (`ClinASO_score`) on the processed Atlas data, following the same pattern as the other competitors (PFRED, sfold, miRanda, OligoWalk).

Requires the `external/clinaso` submodule: `git submodule update --init external/clinaso`.

```
from notebooks.consts import OLIGO_CSV_PROCESSED
import pandas as pd

oligo_data = pd.read_csv(OLIGO_CSV_PROCESSED)
```

```
from notebooks.competitors.clinaso_utils import populate_clinaso
df, features = populate_clinaso(oligo_data)
print(features, df['ClinASO_score'].describe())
```

```
from notebooks.features.feature_extraction import save_feature
for feature in features:
    save_feature(df, feature, version='oligo')
```
